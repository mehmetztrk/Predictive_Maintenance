```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veri setini oku
df = pd.read_csv("../data/ai4i2020.csv")

# İlk 5 satıra göz at
display(df.head())

# Veri setinin yapısı
print(df.shape)
print(df.columns)

# Eksik veri var mı?
print(df.isnull().sum())

# Hedef değişkenin dağılımı (machine failure)
df['Machine failure'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Machine Failure Distribution')
plt.xlabel('Failure (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```
